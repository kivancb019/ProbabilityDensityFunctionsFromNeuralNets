```python



######################################################
## data for crop yields for apples and oranges


## features: temp, rainfall, and humidity

inputs  = np.array([[ 73,  67, 43],
                    [ 91,  88, 64],
                    [ 87, 134, 58],
                    [102,  43, 37],
                    [ 69,  96, 70],
                    [ 73,  67, 43],
                    [ 91,  88, 64],
                    [ 87, 134, 58],
                    [102,  43, 37],
                    [ 69,  96, 70],
                    [ 73,  67, 43],
                    [ 91,  88, 64],
                    [ 87, 134, 58],
                    [102,  43, 37],
                    [ 69,  96, 70]], dtype='float32')

## targets: apples, oranges

targets = np.array([[ 56,  70],
                    [ 81, 101],
                    [119, 133],
                    [ 22,  37],
                    [103, 119],
                    [ 56,  70],
                    [ 81, 101],
                    [119, 133],
                    [ 22,  37],
                    [103, 119],
                    [ 56,  70],
                    [ 81, 101],
                    [119, 133],
                    [ 22,  37],
                    [103, 119]], dtype='float32') 


print(inputs)
print(targets)



```

```python
#######################################################

inputs  = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)

print(inputs)
print(targets)


```

```python
#######################################################
## weights and biases

## 2x3 because 3 targets and 3 features

w = torch.rand(2, 3, requires_grad=True)
b = torch.rand(2,    requires_grad=True)

print(w)
print(b)

#######################################################

preds = model(inputs)

print(preds)
print(targets)


#######################################################
## compute loss

loss = mse(preds, targets)
print(loss)

#######################################################
## derivatives

loss.backward()


print(w)
print(w.grad)
print(b)
print(b.grad)

#######################################################
## Torch needs to reset gradients to zero every iteration

w.grad.zero_()
b.grad.zero_()

print(w.grad)
print(b.grad)

```

```python
######################################################

for i in range(N_EPOCHS):
    preds = model(inputs)
    loss  = mse(preds, targets)
    loss.backward()
    with torch.no_grad():
        w -= w.grad * 1e-5
        b -= b.grad * 1e-5
        w.grad.zero_()
        b.grad.zero_()

        
#######################################################
## calculate loss after training

preds = model(inputs)
loss  = mse(preds, targets)
print(loss)

#######################################################

print(preds)
print(targets)


```

```python
#######################################################
## define dataset

train_ds = TensorDataset(inputs, targets)
print(  train_ds[0:3]  )


```

```python
## define dataloader

train_dl = DataLoader(train_ds, batch_size, shuffle=True)

print(    next(iter(train_dl))   )

```

```python
## define model again using nn

model = nn.Linear(3, 2)    
print(model.weight)
print(model.bias)
```

```python
## define built in optimizer

opt = torch.optim.SGD(   model.parameters(),   lr=1e-5    )
```

```python
loss_fn = F.mse_loss

loss = loss_fn(      model(inputs), targets        )
print(loss)
```

```python
fit(N_EPOCHS, model, loss_fn, opt)

preds = model(inputs)
print(preds)
print(targets)
```

```python


#######################################################

class SimpleNet(nn.Module):
    ## initialize the layers
    def __init__(self):
        super().__init__()
        self.linear1 = nn.Linear(3, 3)
        self.act1    = nn.ReLU()
        self.linear2 = nn.Linear(3, 2)
    
    ## perform inference
    def forward(self, x):
        x = self.linear1(x)
        x = self.act1(x)
        x = self.linear2(x)
        return x
    
#############################################################

model = SimpleNet()
opt = torch.optim.SGD(   model.parameters(), lr=1e-5   )
loss_fn = F.mse_loss
```

```python
fit(1000, model, loss_fn, opt)

preds = model(inputs)
print(preds)
print(targets)
```
